## Task 4: Image to Pencil Sketch with Python

## Importing Libraries


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Reading image

```python
img=cv2.imread('cat.jpg')
org = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.title('Original image')
plt.axis('off')
plt.imshow(org)
```

## RGB image to Grayscale image

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.title('Grayscale image')
plt.axis('off')
plt.imshow(gray, cmap='gray')
```

## Grayscale image to Inverted image

```python
inv=np.invert(gray)
plt.title('Inverted image')
plt.axis('off')
plt.imshow(inv, cmap='gray')
```

## Inverted image to Blurred image

```python
blur = cv2.GaussianBlur(inv, (21, 21), 0)
plt.title('Blurred image')
plt.axis('off')
plt.imshow(blur, cmap='gray')
```

## Pencil Sketch

```python
inv_blur=np.invert(blur)
pencil_sketch = cv2.divide(gray, inv_blur, scale=250)
plt.title('Pencil Sketch')
plt.axis('off')
plt.imshow(pencil_sketch, cmap='gray')
```
